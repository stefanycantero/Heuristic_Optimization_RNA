```python
import numpy as np

# FunciÃ³n para calcular la distancia con la fÃ³rmula de Haversine
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en kilÃ³metros
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

# Crear una matriz de distancias entre todas las ciudades
num_ciudades = len(df_coordenadas)
distancias = np.zeros((num_ciudades, num_ciudades))

for i in range(num_ciudades):
    for j in range(num_ciudades):
        if i != j:  # No calcular la distancia de una ciudad a sÃ­ misma
            distancias[i, j] = haversine(
                df_coordenadas.iloc[i]['Latitud'], df_coordenadas.iloc[i]['Longitud'],
                df_coordenadas.iloc[j]['Latitud'], df_coordenadas.iloc[j]['Longitud']
            )

# Convertir la matriz de distancias a un DataFrame para mejor visualizaciÃ³n
ciudades = df_coordenadas['Ciudad'].values
df_distancias = pd.DataFrame(distancias, index=ciudades, columns=ciudades)

# Verificar las primeras filas de la matriz de distancias
print(df_distancias.head())
```

```python
import numpy as np
import pandas as pd

# ParÃ¡metros del algoritmo
alpha = 1  # Influencia de las feromonas
beta = 2  # Influencia de la visibilidad
evaporacion = 0.5  # Tasa de evaporaciÃ³n de feromonas
Q = 100  # Intensidad de depÃ³sito de feromonas
num_hormigas = 10
num_iteraciones = 100

# FunciÃ³n para calcular las probabilidades de selecciÃ³n
def calcular_probabilidades(ciudad_actual, no_visitadas, feromonas, visibilidad):
    feromonas_ciudad = feromonas.iloc[ciudad_actual, no_visitadas].to_numpy()
    visibilidad_ciudad = visibilidad.iloc[ciudad_actual, no_visitadas].to_numpy()
    probabilidades = (feromonas_ciudad**alpha) * (visibilidad_ciudad**beta)
    return probabilidades / probabilidades.sum()

# Inicializar feromonas y visibilidad
n_ciudades = matriz_costos.shape[0]
feromonas = pd.DataFrame(np.ones((n_ciudades, n_ciudades)), index=matriz_costos.index, columns=matriz_costos.columns)
visibilidad = 1 / matriz_costos.replace(0, np.inf)  # Evitar divisiÃ³n por cero

# Algoritmo de colonia de hormigas
mejor_ruta = None
mejor_costo = float('inf')

for iteracion in range(num_iteraciones):
    rutas = []
    costos = []

    for hormiga in range(num_hormigas):
        ruta = [np.random.randint(0, n_ciudades)]
        no_visitadas = list(set(range(n_ciudades)) - set(ruta))
        costo_ruta = 0

        while no_visitadas:
            ciudad_actual = ruta[-1]
            probabilidades = calcular_probabilidades(ciudad_actual, no_visitadas, feromonas, visibilidad)
            siguiente_ciudad = np.random.choice(no_visitadas, p=probabilidades)
            costo_ruta += matriz_costos.iloc[ciudad_actual, siguiente_ciudad]
            ruta.append(siguiente_ciudad)
            no_visitadas.remove(siguiente_ciudad)

        costo_ruta += matriz_costos.iloc[ruta[-1], ruta[0]]  # Volver a la ciudad inicial
        rutas.append(ruta)
        costos.append(costo_ruta)

        # Actualizar mejor ruta
        if costo_ruta < mejor_costo:
            mejor_ruta = ruta
            mejor_costo = costo_ruta

    # Actualizar feromonas
    feromonas *= (1 - evaporacion)  # EvaporaciÃ³n
    for ruta, costo in zip(rutas, costos):
        for i in range(len(ruta) - 1):
            feromonas.iloc[ruta[i], ruta[i+1]] += Q / costo
            feromonas.iloc[ruta[i+1], ruta[i]] += Q / costo  # ActualizaciÃ³n simÃ©trica

    print(f"IteraciÃ³n {iteracion + 1}: Mejor costo = {mejor_costo}")

# Resultados finales
mejor_ruta_nombres = [matriz_costos.index[ciudad] for ciudad in mejor_ruta]
print("Mejor ruta:", mejor_ruta_nombres)
print("Costo total:", mejor_costo)
```
